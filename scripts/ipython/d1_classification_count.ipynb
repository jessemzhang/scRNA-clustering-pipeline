```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import sys
import os

def load_appropriate_cells(name):
    if 'NoOutliers' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_NoOutliers.txt',dtype=str)
    elif '24_28_30' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_24_28_30.txt',dtype=str)
    elif '22_23' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_22_23.txt',dtype=str)
    elif '22' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_22.txt',dtype=str)
    elif '24' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_24.txt',dtype=str)
    elif '34' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_34.txt',dtype=str)
    elif '35' in temp[0]:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_35.txt',dtype=str)
    else:
        cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_all.txt',dtype=str)
    return cells

# a "0" means group 0 has greater expression of that gene
d1d2_count = np.zeros(326)

# dictionary for mapping all cells
all_cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_all.txt',dtype=str);
dictionary = dict(zip(all_cells, range(0,len(all_cells))))

f = open('/data/jessez/Gene_count_clustering_results/d1d2')
i = 1
for line in f:
    temp = line.split('\t')
    temp2 = temp[6].split(' ')
    
    # load labels
    labels = np.loadtxt('/data/jessez/Gene_count_clustering_results/' + temp[1] + '/labels/labels_' + temp[0])
    
    # determine which cells are involved and convert to index from 0 to 325
    cells = load_appropriate_cells(temp[0])
    ind = np.array(map(lambda x:dictionary[x],cells))
    

    # determine if 0 = d1 or d2
    if temp2[1][0] == '0':  # 0 is d1
        d1d2_count[ind[labels == 0]] += 1
        i += 1
    else:                 # 1 is d1
        d1d2_count[ind[labels == 1]] += 1
        
print i
    
```
