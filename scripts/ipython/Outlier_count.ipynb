```python
%matplotlib inline

# Looking at outliers chosen using various data matrices and various scikit clustering algorithms

import os
import numpy as np
import matplotlib.pyplot as plt

# Create map
cells = np.loadtxt('/data/jessez/Gene_count_datasets/Cells/cells_all.txt',dtype=str);
dictionary = dict(zip(cells, range(0,len(cells))))

# Create counter
outlier_count = np.zeros(len(cells))

def count_outliers_in_dir(dir,counts,dictionary):
    f = lambda x:dictionary[x];
    for i in os.listdir(dir):
        cells_test = np.loadtxt(dir+"/"+i,dtype=str)
        if cells_test.size > 1:
            ind = map(lambda x:dictionary[x],cells_test)
        else:
            ind = f(str(cells_test))
        counts[ind] += 1
    return counts

count_outliers_in_dir('/data/jessez/Gene_count_clustering_results/scikit_kmeans/outliers',outlier_count,dictionary)
count_outliers_in_dir('/data/jessez/Gene_count_clustering_results/scikit_spectral_valiant/outliers',outlier_count,dictionary)
count_outliers_in_dir('/data/jessez/Gene_count_clustering_results/scikit_spectral_rbf/outliers',outlier_count,dictionary)

batch_num = np.loadtxt('/data/jessez/Gene_count_datasets/Batch_nums/batch_num_all.txt')
batches = np.unique(batch_num)

indices = np.array(range(0,len(cells)))

import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

N = len(batches)
iter = 0
for batch in batches:
    ii = (batch_num == batch)
    plt.scatter(indices[ii],outlier_count[ii],c=np.random.rand(3),label=batch)
    iter += 1
plt.legend(loc='best')
plt.xlabel('cell index')
plt.ylabel('frequency of being outlier')
plt.show()
```
